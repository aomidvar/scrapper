```python
import scrapy
import requests
from bs4 import BeautifulSoup

class MySpider(scrapy.Spider):
    name = 'example_spider'
    start_urls = ['http://juhanurmihxlp77nkq76byazcldy2hlmovfu2epvl5ankdibsot4csyd.onion/']

    def parse(self, response):
        title = response.css('title::text').get()
        description = response.xpath('//meta[@name="description"]/@content').get()

        yield {
            'Title': title,
            'Description': description,
        }

# Execute the spider
from scrapy.crawler import CrawlerProcess

process = CrawlerProcess()
process.crawl(MySpider)
process.start()

# Display the scraped data
data = process.crawl(MySpider)
for item in data:
    print(item)
```

```python

import concurrent.futures
import requests
from bs4 import BeautifulSoup
import json

def scrape_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            title = soup.title.string if soup.title else "No Title"
            description = soup.find('meta', attrs={'name': 'description'})
            description = description['content'] if description else "No Description"
            return {'url': url, 'title': title, 'description': description}
    except Exception as e:
        print(f"Error scraping {url}: {str(e)}")
    return None

def main():
    # Load your list of URLs from a JSON file or any other data source.
    urls = [...]

    # Set the number of URLs to scrape in each thread.
    chunk_size = 50

    scraped_data = []

    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
        for i in range(0, len(urls), chunk_size):
            chunk = urls[i:i+chunk_size]
            results = list(executor.map(scrape_url, chunk))
            scraped_data.extend([r for r in results if r is not None])

    # Store the scraped data in a JSON file or database.
    with open('scraped_data.json', 'w') as file:
        json.dump(scraped_data, file, indent=4)

if __name__ == "__main__":
    main()
```
